```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Ellipse
import numpy as np
from itertools import combinations
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'

# --- Source data from Anhang F ---
# Target totals: (a)=26, (b)=19, (c)=8, (d)=19

sources = [
    "Adeosun & Oro (2023)",
    "Agal et al. (2025)",
    "Arimond et al. (2024)",
    "Ashraf et al. (2024)",
    "Batrancea et al. (2025)",
    "Bosancic et al. (2024)",
    "Charkhestani et al. (2026)",
    "Clare et al. (2021a)",
    "Clare et al. (2025)",
    "Collin-Dufresne et al. (2020)",
    "Denekew et al. (2025)",
    "Dua et al. (2021)",
    "Erlwein-Sayer et al. (2020)",
    "Fons et al. (2018)",
    "Goodarzi & Meinerding (2023)",
    "Ha Jeon et al. (2023)",
    "Jung et al. (2022)",
    "Ke et al. (2026)",
    "Khemlichi et al. (2025)",
    "Kieren & Weber (2022)",
    "Kijkarncharoensin (2025)",
    "Kumar et al. (2024)",
    "Leggio & Lien (2024)",
    "Li & Mulvey (2021)",
    "Li et al. (2025)",
    "Li (2020)",
    "Mari & Baldassari (2022)",
    "Mueller-Glissmann & Ferrario (2024)",
    "Nest et al. (2023)",
    "Oliveira & Costa (2025)",
    "Oprisor & Kwon (2021)",
    "Pallotta & Ciciretti (2024)",
    "Pun et al. (2023)",
    "Rahaman (2025)",
    "Rani et al. (2025)",
    "Salotra et al. (2026)",
    "Shi (2024)",
    "Shu et al. (2024a)",
    "Shu & Mulvey (2025)",
    "Shu et al. (2024b)",
    "Šikić (2025)",
    "Song et al. (2025)",
    "Soni (2024)",
    "Strauss (2025)",
    "Suvak et al. (2025)",
    "Taguchi et al. (2022)",
    "Taneva-Angelova & Granchev (2025)",
    "Wang et al. (2020)",
    "Wooh & Cho (2025)",
    "Zhu et al. (2025)",
]

# Binary membership: (a), (b), (c), (d)
data = np.array([
    #                                  (a) (b) (c) (d)
    [0, 0, 0, 1],  #  0 Adeosun & Oro (2023)
    [0, 1, 0, 1],  #  1 Agal et al. (2025)
    [0, 1, 0, 1],  #  2 Arimond et al. (2024)
    [0, 1, 0, 0],  #  3 Ashraf et al. (2024)
    [0, 1, 0, 1],  #  4 Batrancea et al. (2025)
    [1, 1, 0, 1],  #  5 Bosancic et al. (2024)
    [0, 1, 0, 0],  #  6 Charkhestani et al. (2026)
    [0, 0, 1, 0],  #  7 Clare et al. (2021a)
    [0, 0, 1, 0],  #  8 Clare et al. (2025)
    [1, 0, 0, 0],  #  9 Collin-Dufresne et al. (2020)
    [0, 1, 0, 0],  # 10 Denekew et al. (2025)
    [1, 0, 0, 0],  # 11 Dua et al. (2021)
    [1, 0, 0, 0],  # 12 Erlwein-Sayer et al. (2020)
    [1, 0, 0, 0],  # 13 Fons et al. (2018)
    [1, 0, 0, 0],  # 14 Goodarzi & Meinerding (2023)
    [1, 1, 0, 1],  # 15 Ha Jeon et al. (2023)
    [1, 0, 1, 0],  # 16 Jung et al. (2022)
    [0, 0, 0, 1],  # 17 Ke et al. (2026)
    [0, 1, 0, 0],  # 18 Khemlichi et al. (2025)
    [0, 0, 1, 0],  # 19 Kieren & Weber (2022)
    [0, 1, 0, 1],  # 20 Kijkarncharoensin (2025)
    [1, 0, 0, 0],  # 21 Kumar et al. (2024)
    [0, 0, 1, 0],  # 22 Leggio & Lien (2024)
    [0, 0, 1, 0],  # 23 Li & Mulvey (2021)
    [1, 0, 0, 0],  # 24 Li et al. (2025)
    [1, 1, 0, 1],  # 25 Li (2020)
    [1, 0, 0, 0],  # 26 Mari & Baldassari (2022)
    [1, 0, 0, 0],  # 27 Mueller-Glissmann & Ferrario (2024)
    [1, 1, 0, 1],  # 28 Nest et al. (2023)
    [1, 1, 0, 1],  # 29 Oliveira & Costa (2025)
    [1, 0, 0, 1],  # 30 Oprisor & Kwon (2021)
    [0, 1, 0, 0],  # 31 Pallotta & Ciciretti (2024)
    [1, 0, 0, 0],  # 32 Pun et al. (2023)
    [0, 1, 0, 1],  # 33 Rahaman (2025)
    [0, 0, 0, 1],  # 34 Rani et al. (2025)
    [1, 1, 1, 1],  # 35 Salotra et al. (2026)
    [1, 0, 0, 0],  # 36 Shi (2024)
    [1, 0, 0, 0],  # 37 Shu et al. (2024a)
    [1, 0, 0, 1],  # 38 Shu & Mulvey (2025)
    [1, 0, 0, 1],  # 39 Shu et al. (2024b)
    [1, 0, 0, 0],  # 40 Šikić (2025)
    [0, 1, 0, 0],  # 41 Song et al. (2025)
    [0, 0, 1, 0],  # 42 Soni (2024)
    [1, 0, 0, 0],  # 43 Strauss (2025)
    [1, 0, 0, 0],  # 44 Suvak et al. (2025)
    [0, 1, 0, 0],  # 45 Taguchi et al. (2022)
    [0, 0, 0, 1],  # 46 Taneva-Angelova & Granchev (2025)
    [1, 0, 0, 0],  # 47 Wang et al. (2020)
    [0, 1, 0, 0],  # 48 Wooh & Cho (2025)
    [0, 0, 0, 1],  # 49 Zhu et al. (2025)
])

# --- Verify totals ---
totals = [data[:, j].sum() for j in range(4)]
target = [26, 19, 8, 19]
print(f"Computed totals: (a)={totals[0]}, (b)={totals[1]}, (c)={totals[2]}, (d)={totals[3]}")
print(f"Target totals:   (a)={target[0]}, (b)={target[1]}, (c)={target[2]}, (d)={target[3]}")
assert totals == target, f"MISMATCH! {totals} != {target}"
print("✓ Totals match Anhang F.\n")

# --- Build sets ---
dim_names = ["a", "b", "c", "d"]
dim_labels = [
    "(a) Ökonometrische\nRegime-Identifikation &\nAsset-Allokation",
    "(b) Deep Learning\nfür Finanzzeitreihen",
    "(c) SORR &\nEntnahmephase",
    "(d) Hybride\nModellansätze",
]

sets = {}
for i, name in enumerate(dim_names):
    sets[name] = set(j for j in range(len(sources)) if data[j, i] > 0)

# --- Helper: exclusive region for a given combination ---
def exclusive(included_names):
    """Papers in ALL of included_names and NONE of the others."""
    result = sets[included_names[0]]
    for name in included_names[1:]:
        result = result & sets[name]
    for name in dim_names:
        if name not in included_names:
            result = result - sets[name]
    return result

# --- Exclusive counts for only-one-set regions ---
only = {name: exclusive([name]) for name in dim_names}

# --- Print statistics ---
print("Set sizes:")
for name in dim_names:
    print(f"  {name}: n = {len(sets[name])}")

print("\nExclusive single-set regions:")
for name in dim_names:
    print(f"  only_{name}: {len(only[name])}")

print("\nExclusive pairwise intersections:")
for c in combinations(dim_names, 2):
    ex = exclusive(list(c))
    print(f"  {c[0]}∩{c[1]} excl.: {len(ex)}  {[sources[i] for i in ex]}")

print("\nExclusive triple intersections:")
for c in combinations(dim_names, 3):
    ex = exclusive(list(c))
    print(f"  {c[0]}∩{c[1]}∩{c[2]} excl.: {len(ex)}  {[sources[i] for i in ex]}")

quad = sets["a"] & sets["b"] & sets["c"] & sets["d"]
print(f"\nQuadruple intersection (a∩b∩c∩d): {len(quad)}")
for i in quad:
    print(f"  → {sources[i]}")

# --- Sanity check: all exclusive regions must sum to 50 ---
total_check = sum(len(only[n]) for n in dim_names)
for c in combinations(dim_names, 2):
    total_check += len(exclusive(list(c)))
for c in combinations(dim_names, 3):
    total_check += len(exclusive(list(c)))
total_check += len(quad)
print(f"\nSum of all exclusive regions: {total_check}")
assert total_check == len(sources), f"MISMATCH! {total_check} != {len(sources)}"
print("Sum matches total number of sources.\n")

# =====================================================================
# VENN DIAGRAM
# =====================================================================
fig, ax = plt.subplots(figsize=(12, 10))
ax.set_xlim(-6.5, 6.5)
ax.set_ylim(-5.5, 4.8)
ax.set_aspect('equal')
ax.axis('off')

# 4 overlapping ellipses (tilted arrangement)
ellipse_params = [
    (-1.5,  1.0, 6.5, 4.2,  45, "#2E86AB"),   # (a) top-left
    ( 1.5,  1.0, 6.5, 4.2, -45, "#A23B72"),   # (b) top-right
    (-1.5, -1.0, 6.5, 4.2, -45, "#F18F01"),   # (c) bottom-left
    ( 1.5, -1.0, 6.5, 4.2,  45, "#3B8132"),   # (d) bottom-right
]

label_positions = [
    (-4.8,  3.8),   # (a) top-left
    ( 4.8,  3.8),   # (b) top-right
    (-4.8, -3.8),   # (c) bottom-left
    ( 4.8, -3.8),   # (d) bottom-right
]

label_ha = ["right", "left", "right", "left"]

for i, (cx, cy, w, h, angle, color) in enumerate(ellipse_params):
    e = Ellipse((cx, cy), w, h, angle=angle,
                facecolor=color, alpha=0.18, edgecolor=color,
                linewidth=2.5, linestyle='-')
    ax.add_patch(e)

# Dimension labels with counts
for i, (lx, ly) in enumerate(label_positions):
    n = len(sets[dim_names[i]])
    ax.text(lx, ly, f"{dim_labels[i]}\n(n = {n})",
            fontsize=9.5, fontweight="bold",
            ha=label_ha[i], va="center",
            color=ellipse_params[i][5],
            bbox=dict(boxstyle="round,pad=0.3", facecolor="white",
                      edgecolor=ellipse_params[i][5], alpha=0.9))

# --- Place EXCLUSIVE counts ---

# Only-one-set positions
only_positions = {
    "a": ( -2.0, 2.8),   # top-left, only inside a
    "b": ( 2.0, 2.8),   # top-right, only inside b
    "c": (-2.0, -2.8),   # bottom-left, only inside c
    "d": ( 2.0, -2.8),   # bottom-right, only inside d
}

for name, (px, py) in only_positions.items():
    n = len(only[name])
    if n > 0:
        ax.text(px, py, str(n), fontsize=13, fontweight="bold",
                ha="center", va="center", color="#333",
                bbox=dict(boxstyle="round,pad=0.2", facecolor="white",
                          edgecolor="#999", alpha=0.8))

# Pairwise EXCLUSIVE positions
pair_positions = {
    ("a", "b"): ( 0.0,  3.0),   # top center, a∩b ohne c,d
    ("a", "c"): (-2.8, -0.5),   # left, a∩c ohne b,d
    #("a", "d"): ( -1.0, 0.6),   # left of center, a∩d ohne b,c; deleted — see annotation below
    ("b", "c"): ( 0.6,  0.0),   # right of center, b∩c ohne a,d  (currently 0)
    ("b", "d"): ( 2.8, -0.5),   # right, b∩d ohne a,c
    ("c", "d"): ( 0.0, -3.0),   # bottom center, c∩d ohne a,b
}

for (s1, s2), (px, py) in pair_positions.items():
    n = len(exclusive([s1, s2]))
    if n > 0:
        ax.text(px, py, str(n), fontsize=13, fontweight="bold",
                ha="center", va="center", color="#333",
                bbox=dict(boxstyle="round,pad=0.2", facecolor="white",
                          edgecolor="#999", alpha=0.8))

# Triple EXCLUSIVE intersections
triple_positions = {
    ("a", "b", "c"): (-0.6,  0.8),   # upper-left of center
    ("a", "b", "d"): ( 0.6, 0.7),   # upper-right of center
    ("a", "c", "d"): (-0.6, -0.8),   # lower-left of center
    ("b", "c", "d"): ( 0.6, -0.8),   # lower-right of center
}

for (s1, s2, s3), (px, py) in triple_positions.items():
    n = len(exclusive([s1, s2, s3]))
    if n > 0:
        ax.text(px, py, str(n), fontsize=12, fontweight="bold",
                ha="center", va="center", color="#555",
                bbox=dict(boxstyle="round,pad=0.2", facecolor="#FFFDE7",
                          edgecolor="#999", alpha=0.85))

# --- Quadruple intersection: CENTER with emphasis ---
quad_n = len(quad)
center_color = "#D32F2F"

circle = plt.Circle((0, 0), 0.4,
                     facecolor="#FFCDD2",
                     edgecolor=center_color, linewidth=2.5,
                     linestyle="--", zorder=5)
ax.add_patch(circle)
ax.text(0, 0, str(quad_n), fontsize=13, fontweight="bold",
        ha="center", va="center", color=center_color, zorder=6)

# --- Annotation ---
ax.annotate(
    "Forschungslücke:\nSalotra & Pinsky (2026) deckt\n"
    "formal alle vier Dimensionen\n"
    "ab, weicht jedoch vom\n"
    "Forschungsziel dieser Arbeit ab",
    xy=(-0.4, 0),
    xytext=(-5.5, 0.0),
    fontsize=8.5, fontweight="bold", color=center_color,
    ha="center", va="center",
    bbox=dict(boxstyle="round,pad=0.5", facecolor="#FFF3E0",
              edgecolor=center_color, linewidth=1.5),
    arrowprops=dict(arrowstyle="-|>", color=center_color,
                    lw=1.8, connectionstyle="arc3,rad=-0.2"),
    zorder=7
)
# --- Annotation for a∩d exclusive (not visible as distinct region) ---
ax.annotate(
    "a ∩ d exkl.: 3\n"
    "Oprisor & Kwon (2021)\n"
    "Shu & Mulvey (2025)\n"
    "Shu et al. (2024b)",
    xy=(0.5, 0.0),          # Pfeilspitze zeigt auf die a∩d-Gegend
    xytext=(5.5, -1.5),     # Textbox links außerhalb
    fontsize=7.5, color="#555",
    ha="center", va="center",
    bbox=dict(boxstyle="round,pad=0.4", facecolor="#E8F5E9",
              edgecolor="#777", linewidth=1.0),
    arrowprops=dict(arrowstyle="-|>", color="#777",
                    lw=1.5, connectionstyle="arc3,rad=0.2"),
    zorder=7
)

plt.tight_layout()
plt.savefig('../../assets/concept_matrix_venn.png', dpi=300,
            bbox_inches="tight", facecolor="white")
plt.close()
print("\nVenn diagram saved.")
```
