```python
# S&P 500 mit NBER-Rezessionsphasen

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime

# S&P 500 Daten laden
sp500 = yf.download("^GSPC", start="1990-01-01", end="2026-04-18")
sp500 = sp500["Close"].squeeze()

# NBER-Rezessionsphasen (offizielle Datierung)
recessions = [
    ("1990-07-01", "1991-03-01"),  # Early 1990s Recession
    ("2001-03-01", "2001-11-01"),  # Dotcom-Rezession
    ("2007-12-01", "2009-06-01"),  # Globale Finanzkrise
    ("2020-02-01", "2020-04-01"),  # COVID-19-Rezession
]

# Chart erstellen
fig, ax = plt.subplots(figsize=(14, 6))

# S&P 500 Kursverlauf
ax.plot(sp500.index, sp500.values, color="black", linewidth=0.8, label="S&P 500")

# Rezessionsphasen als graue Flächen
for start, end in recessions:
    ax.axvspan(
        pd.Timestamp(start), 
        pd.Timestamp(end), 
        color="lightcoral", 
        alpha=0.35, 
        label="NBER-Rezession" if start == recessions[0][0] else None
    )

# Formatierung
ax.set_xlabel("Jahr", fontsize=11)
ax.set_ylabel("S&P 500 (Schlusskurs in USD)", fontsize=11)
ax.set_title("S&P 500 mit NBER-Rezessionsphasen (1990–2025)", fontsize=13, fontweight="bold")
ax.legend(loc="upper left", fontsize=10)
ax.xaxis.set_major_locator(mdates.YearLocator(5))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y"))
ax.set_xlim(pd.Timestamp("1990-01-01"), pd.Timestamp("2026-04-18"))
ax.grid(axis="y", alpha=0.3)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()
plt.savefig("../assets/s&p500_nber_rezessionen.png", dpi=300, bbox_inches="tight")
plt.show()

print(f"Datenpunkte: {len(sp500)}")
print(f"Zeitraum: {sp500.index[0].strftime('%Y-%m-%d')} bis {sp500.index[-1].strftime('%Y-%m-%d')}")
```
