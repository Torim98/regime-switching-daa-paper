```python
# --- Central Config ---
import sys; sys.path.insert(0, "../config")
from config_loader import cfg

if cfg.dependencies.install_on_run:
    # Abhängigkeiten installieren
    import subprocess, sys
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "-e", ".."])
    print("Dependencies installiert (aus pyproject.toml).")
else:
    print("Dependency-Installation übersprungen (dependencies.install_on_run = false).")

# --- nbstripout: Notebook-Outputs automatisch aus Git-Commits ausschließen ---
if cfg.dependencies.nbstripout:
    import subprocess
    try:
        subprocess.run(["nbstripout", "--version"], capture_output=True, check=True)
    except (FileNotFoundError, subprocess.CalledProcessError):
        %pip install -q nbstripout
    result = subprocess.run(
        ["nbstripout", "--install", "--attributes", "../.gitattributes"],
        capture_output=True, text=True
    )
    if result.returncode == 0:
        print("nbstripout Git-Filter aktiviert (Outputs werden beim Commit automatisch gestripped).")
    else:
        print(f"nbstripout Setup fehlgeschlagen: {result.stderr}")
else:
    print("nbstripout übersprungen (dependencies.nbstripout = false).")
```
